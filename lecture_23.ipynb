### lecture 23 - Trend Changes

```python
import pandas as pd
from prophet import Prophet
```

```python
df = pd.read_csv('./dataset/HospitalityEmployees.csv')
```

```python
df.columns = ['ds','y']
```

```python
df['ds'] = pd.to_datetime(df['ds'])
```

```python
df.plot(x='ds',y='y',figsize=(12,8))
```

```python
m = Prophet()
```

```python
m.fit(df)
```

```python
future = m.make_future_dataframe(periods=12,freq='MS')
```

```python
forecast = m.predict(future)
```

```python
from prophet.plot import add_changepoints_to_plot
```

```python
fig = m.plot(forecast)
a = add_changepoints_to_plot(fig.gca(),m,forecast)
```

from the last years we have a very stady trend
