### rolling and expanding

```python
import pandas as pd
```

```python
df = pd.read_csv('./dataset/starbucks.csv',index_col='Date',parse_dates=True)
```

```python
df.head()
```

#### rolling

```python
df['Close'].plot(figsize=(12,5))
```

```python
# window decides how large of a section we want to look at the time
df.rolling(window=7).mean() 
```

the first 6(window-1) values are NaN because we do not have yet enough data to actually take the average

```python
df['Close'].plot(figsize=(12,5))
df.rolling(window=7).mean()['Close'].plot()
```

```python
df['Close'].plot(figsize=(12,5))
df.rolling(window=15).mean()['Close'].plot() 
```

```python
df['Close'].plot(figsize=(12,5))
df.rolling(window=30).mean()['Close'].plot() # average price for 30 days window
```

```python
df['Close'].plot(figsize=(12,5))
df.rolling(window=60).mean()['Close'].plot() 
```

```python
df['Close: 30 day mean'] = df['Close'].rolling(window=30).mean()
```

```python
df.head() # first 29 values are going to be empty
```

```python
df[['Close','Close: 30 day mean']].plot(figsize=(12,5))
```

#### expanding

```python
df['Close'].expanding().mean().plot(figsize=(12,5))
```

the last value in the plot at '2019-01' is going to be the mean of all the rows in the dataset
